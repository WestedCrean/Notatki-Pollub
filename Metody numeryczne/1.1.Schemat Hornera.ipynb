# Schemat Hornera

Niech $P_n$ oznacza zbiór wielomianów stopnia $\lt n$

```python
import numpy as np

n = 14

wspolczynniki = np.random.rand(n)

def Horner(n,a,c):
    # n - stopien wielomianu
    # a - tablica wspolczynnikow
    # c - 'x' dla ktorego obliczyc wielomian
    
    b = a[0]
    for element in a:
        b = b * c + element
    return b

obliczony = Horner(n,wspolczynniki,3.14)
print(obliczony)
```
